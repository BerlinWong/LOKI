```python
# Import dependencies 导入依赖
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Define the sigmoid function 定义Sigmoid函数
# The point of this is to let our output value range in (0,1) 这样做的意义在于让我们的输出值范围在(0,1)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# 画出sigmoid函数
x_list = np.arange(-10, 10, 0.01)


def draw_the_sigmoid_function_line(z):
    function_y_values = []

    for i in z:
        function_y_values.append(sigmoid(i))

    plt.title('Sigmoid Function')
    plt.xlabel('x')
    plt.ylabel('σ(x)')
    plt.plot(z, function_y_values)
    plt.grid()
    plt.show()


draw_the_sigmoid_function_line(x_list)
```

```python
class Neuron:
    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def feedforward(self, inputs):
        total = np.dot(inputs, self.weights) + self.bias
        return sigmoid(total)


weights = np.array([0, 1])
bias = 4
n = Neuron(weights, bias)

x = np.array([2, 3])

print(n.feedforward(x))
```
